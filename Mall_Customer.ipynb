```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
customer_data = pd.read_csv('/content/Mall_Customers.csv')
```

```python
customer_data.head()
```

```python
customer_data.info()
```

```python
customer_data.isnull().sum()
```

```python
x = customer_data.drop(columns = ['CustomerID','Gender','Age'], axis=1).values
```

### *Visualize The Data Points*

```python
plt.figure(figsize = (15,8))
sns.scatterplot(x.ravel())
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score')
plt.show()
```

### *Find the K value using the Elbow Method*

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=2)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

plt.figure(figsize = (15,8))
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('wcss')
plt.show()
```

### *Training the K-means Algorithm on the training dataset*

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 0)
y = kmeans.fit_predict(x)
```

### *Centroid Points*

```python
kmeans.cluster_centers_
```

### *Visualize The Clusters Formed*

```python
plt.figure(figsize = (15,8))
plt.scatter(x[y==0,0], x[y==0,1], s=50, c='red', label='cluster 1')
plt.scatter(x[y==1,0], x[y==1,1], s=50, c='pink', label='cluster 2')
plt.scatter(x[y==2,0], x[y==2,1], s=50, c='yellow', label='cluster 3')
plt.scatter(x[y==3,0], x[y==3,1], s=50, c='green', label='cluster 4')
plt.scatter(x[y==4,0], x[y==4,1], s=50, c='orange', label='cluster 5')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=200, c = 'black', label = 'Clusters' )
plt.title('Customer Groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
